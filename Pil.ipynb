```python
altura = 256
largura = 256
rgb = (0, 0, 0)#preto
nova = Image.new('RGB', (altura,largura), rgb)
```

```python
nova
```

```python
nova.putpixel((5,5), (255,255,255))
```

```python
nova
```

```python
def quadrado(im, x, y, rgb):
    for i in range(x, x + 15):
        for j in range(y, y + 15):
            im.putpixel((i,j), rgb)
```

```python
vermelha = (255,0,0)
```

```python
verde = (0,255,0)
```

```python
quadrado(nova, 5, 5, vermelha)
```
