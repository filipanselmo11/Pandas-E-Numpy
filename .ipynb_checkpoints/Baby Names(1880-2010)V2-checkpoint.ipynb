```python
anos = range(1880, 2011)
partes = []
colunas = ['nome', 'sexo', 'nascimentos']

for ano in anos:
    caminho = 'yob1880.txt'
    frame = pd.read_csv(caminho, names=colunas)
    
    frame['ano'] = ano
    partes.append(frame)
nomes = pd.concat(partes, ignore_index=True)
```
